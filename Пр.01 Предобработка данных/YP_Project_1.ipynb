### Обработка пропусков

```python
def FillnaTotalIncome(param):
    median_data_ti = data[data['income_type'] == param]['total_income'].median()
    data.loc[data['income_type']==param,'total_income'] = data.loc[data['income_type']==param,'total_income'].fillna(median_data_ti)
    return data
    
#заполняем пропуске в столбце total_income

#смотрим, сколько всего у нас пропусков в total_income
print('Всего пропусков в total_income\n',data[data['total_income'].isnull()].count()) 

#выясняем, как пропуски в total_income распределены по типам занятости
print('\nРаспределение пропусков по типам занятости\n',data['total_income'].isnull().groupby(data['income_type']).sum().sort_values()) 

#список типов занятости, у которых есть пропуски в total_income
NaN_income_type=['госслужащий','компаньон','пенсионер','предприниматель','сотрудник'] 

#пропуски заполним медианными значениями. 
#считаем медиану, а не среднее, потому что большой разброс между максимальным и минимальным значением дохода
#поскольку у разных типов занятости разные уровни дохода, рассчитаем медиану для каждого типа занятости
for q in NaN_income_type:
    data = FillnaTotalIncome(q)

#проверяем, остались ли пустые значения в total_income
print('\nПропущенных значений в total_income после заполнения:','\n',data[data['total_income'].isnull()].count()) 
```

### Вывод

Заполнили пропуски в столбце total_income медианными значениями, рассчитанными для каждого типа занятости.

### Лемматизация

```python
def LemmatizeIt(purpose):
    lemmas = m.lemmatize(purpose)
    if 'жилье' in lemmas or 'недвижимость' in lemmas:
        return 'недвижимость'
    if 'автомобиль' in lemmas:
        return 'автомобиль'
    if 'образование' in lemmas:
        return 'образование'
    return 'свадьба'

#смотрим, какие варианты вообще есть
print('Цели кредита\n',data['purpose'].unique()) 

#категоризируем цель кредита после лемматизации
data['purpose'] = data['purpose'].apply(LemmatizeIt)
print('\nЦели кредита после лемматизации и категоризации\n',data['purpose'].unique()) #проверили
```

### Вывод

Лемматизировали цели кредита и заодно категоризировали их, сократив до 4 категорий.
Лемматизировали - значит брали поочередно в каждой табличной записи цель кредита и приводили каждое слово в ней к словарной форме (лемме). 

### Категоризация данных

```python
def FamilyStatusCategorizer (family_status):
    family_status_single = ['Не женат / не замужем','в разводе','вдовец / вдова','гражданский брак']
    if family_status in family_status_single:
        return 'не женат / не замужем'
    return 'женат/замужем'

def FamilyStatusIdCategorizer (family_status_id):
    family_status_id_single = [4,3,2,1]
    if family_status_id in family_status_id_single:
        return 0
    return 1

def TotalIncomeGroup(total_income):
    if total_income < 100000:
        return 'менее 100 т.р.'
    if total_income < 200000:
        return 'от 100 до 200 т.р.'
    if total_income < 300000:
        return 'от 200 до 300 т.р.'
    return 'более 300 т.р.'

def ChildrenGroup(children):
    if children == 0:
        return 0
    return 1

#категоризируем семейный статус в две категории: женат/замужем и не женат/не замужем 
#все остальные нюансы банку думаю не интересны, поскольку юридически никак не оформлены
data['family_status'] = data['family_status'].apply(FamilyStatusCategorizer)

#также категоризируем id семеного статуса (1 - женат/замужем, 0 - не женат/не замужем)
data['family_status_id'] = data['family_status_id'].apply(FamilyStatusIdCategorizer)
print('Семейный статус\n',data['family_status'].value_counts()) #проверили
print('\nID cемейного статуса\n',data['family_status_id'].value_counts()) #проверили

#категоризируем уровень дохода в месяц (менее 100, от 100 до 200, от 200 до 300, более 300)
data['total_income_group'] = data['total_income'].apply(TotalIncomeGroup)
print('\nУровень дохода в месяц\n',data['total_income_group'].value_counts().sort_values()) #проверили

#категоризируем количество детей в наличие детей: есть(1)/нет(0), поскольку нам интересно наличие, а не количество
print('\nКоличество детей\n',data['children'].value_counts()) #смотрим, какие есть значения и сколько их
#ошибочные записи -1 (47) и 20 (76) - будем считать, что это 1 и 2 - их все равно мало и они нам погоду не сделают, 
#поскольку бездетных все равно в два раза больше
data['children_group'] = data['children'].apply(ChildrenGroup)
print('\nНаличие детей\n',data['children_group'].value_counts()) #проверили
```

Можно было бы вынести семейный статус в отдельный словарь, но у нас всего две категории - смысл?

### Шаг 3. Ответьте на вопросы

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
#делаем сводную таблицу по числу клиентов
children_data_pivot_count = analys_data.pivot_table(index='children_group', values='debt', aggfunc='count') 
#делаем сводную таблицу по числу должников
children_data_pivot_sum = analys_data.pivot_table(index='children_group', values='debt', aggfunc='sum')
#объединяем
children_data_pivot = children_data_pivot_count.merge(children_data_pivot_sum, on='children_group', how='left')
#переобзываем столбцы
children_data_pivot.set_axis(['clients','debitors'],axis='columns',inplace=True)
#высчитываем доли должников
children_data_pivot['ratio,%'] = round((children_data_pivot['debitors'] / children_data_pivot['clients'])*100,1)
#сортируем по доле должников по возрастанию и смотрим на нашу таблицу
children_data_pivot.sort_values(by='ratio,%').head()
```

### Вывод

Люди без детей почти вдвое чаще людей с детьми берут кредиты, при этом особой зависимости между наличием детей и возвратом кредита в срок не наблюдается - должники есть и в той и в другой социальной группе, при этом разница в доли должников не более 1.7%.
